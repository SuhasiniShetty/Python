```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) # adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:/Users/s.nagraj.shetty/Downloads/movies.csv')
df.head()
```

```python
# Finding missing data if any

for col in df.columns:
    gap_prcnt = np.round(np.mean(df[col].isnull()),1)
    print('{} - {}%'.format(col,gap_prcnt))
```

```python
df.head(10)
```

```python
# data types of our columns

df.dtypes
```

```python
# Create Column yearcorrect using released column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(str)
df.head()
```

```python
# Sort the dataset by 'Gross' in ascending 

df = df.sort_values(by=['gross'],inplace = False, ascending = False)
df.head()
```

```python
# setting the data set view to display all columns and rows (default is 20 rows)

pd.set_option('display.max_rows',None)
```

```python
# Hypothesis_1
# Budget highly correlated to Gross
# Scatter plot to compare

df = df.dropna()
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget vs gross using regplot in Seaborn

sns.regplot(x="budget", y="gross", data=df, scatter_kws = {"color" : "red"}, line_kws = {"color" : "blue"})
```

```python
# Types of correlation
# pearson, kendall, spearman

df.corr(method = "pearson") 
```

```python
# Hypothesis_1 was accurate as we have high correlation between Budget and Gross as seen in the table above
```

```python
# Visualizing our correlation matrix using seaborn heatmap

correlation_matrix = df.corr(method = "pearson")

sns.heatmap(correlation_matrix,annot = True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Catergorizing the "Object" fields in the data frame

df_num = df

for col_name in df_num.columns:
    if (df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
df_num.head()
```

```python
# Visualizing our correlation matrix using seaborn heatmap with the numerized dataframe

correlation_matrix = df_num.corr(method = "pearson")

sns.heatmap(correlation_matrix,annot = True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# unstacking the correlation matrix

corr_mat = df_num.corr()
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
# Sorting the unstacked correlation matrix

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# finding featured with a correlation value greater than 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
